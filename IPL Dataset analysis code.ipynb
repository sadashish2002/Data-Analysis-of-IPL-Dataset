```python
#Find out No of Toss Wins w.r.t. Each Team
ipl['toss_winner'].value_counts()
```

```python

#Extracting the records where a team won batting first

batting_first=ipl[ipl['win_by_runs']!=0]
batting_first.head()
```

```python
#creating plot of toss decision
plt.figure(figsize=(10,6))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.show()
```

```python
#No of WIns w.r.t. Rach Team After Batting First
batting_first['winner'].value_counts()
```

```python
list(batting_first['winner'].value_counts().keys())
```

```python
#Top 3 team who wins after batting first
a =batting_first['winner'].value_counts()[0:3]
a
```

```python
#Extract name of top 3 team who wins after batting first
b =list(batting_first['winner'].value_counts().keys()[0:3])
b
```

```python
#Creating barplot for Top 3 Teamsn Wins After Batting First
plt.figure(figsize=(10,6))
c = ['blue','yellow','green']
plt.bar(b,a,color=c)
plt.xlabel("Name of teams")
plt.ylabel("NO of matches wins after batting first")
plt.title("Top 3 team Statistics")

plt.show()
```

```python
data=list(batting_first['winner'].value_counts())
data
```

```python
Label=list(batting_first['winner'].value_counts().keys())
Label
```

```python
#Making a pie chart
plt.figure(figsize=(8,8))
plt.pie(data ,labels =Label,autopct ='%1.1f%%',shadow = True)
plt.title('Winnig Percentage of IPL teams after Batting first')
plt.show()
```

```python
#Extracting those  Records where a Team has won after Batting secnd
batting_second = ipl[ipl['win_by_wickets']!=0]
batting_second.head()
```

```python
e =batting_second['winner'].value_counts()
e
```

```python
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.title('win by Wickets')
plt.show()
```

```python
#extract top 3 team who wins after second batting
f =list(batting_second['winner'].value_counts()[0:3])
f
```

```python
g =list(batting_second['winner'].value_counts()[0:3].keys())
g
```

```python
#Making a bar plot for top 3  teams with win after batting second
plt.figure(figsize=(7,7))
c = ['purple','blue','red']
plt.bar(g,f,color=c)
plt.show()
```

```python
#Making Pie chart for wins after batting second
data = list(batting_second['winner'].value_counts())
label= list(batting_second['winner'].value_counts().keys())
plt.figure(figsize=(7,7))
plt.pie(data,labels =label,autopct ='%1.1f%%',shadow =True)
plt.title('Teams wins % Statistics after 2nd batting')
plt.show()
```
